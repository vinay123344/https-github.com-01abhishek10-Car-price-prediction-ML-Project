# 1. Importing the libraries

```python
import pandas as pd
import numpy as np
```

```python
xls = pd.ExcelFile('Hash-Analytic-Python-Analytics-Problem-case-study-1.xlsx')
df1=pd.read_excel(xls,'Existing employees')
df2=pd.read_excel(xls,'Employees who have left')
```

```python
df1.head()
```

```python
df2.head()
```

# 2. Combining data of employees who left and existing employees in order to study the status visually

```python
df1['status']=0
df2['status']=1
frames=[df1,df2]
df=pd.concat(frames, axis=0, ignore_index=True)
df=df.sort_values(by='Emp ID', ascending=True)
```

```python
df
```

# 3. Encoding categorical data

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
```

```python
enc=LabelEncoder()
```

```python
count=0
for i in df1.columns[0:]:
    if df1[i].dtype=='object':
        enc.fit(df1[i])
        df1[i]=enc.transform(df1[i])
        count+=1
```

```python
count
```

```python
enc1 = OneHotEncoder(handle_unknown='ignore')
```

```python
enc_df1 = pd.DataFrame(enc1.fit_transform(df[['dept']]).toarray())
enc_df2 = pd.DataFrame(enc1.fit_transform(df[['salary']]).toarray())
```

```python
enc_df1
```

```python
enc_df1=enc_df1.drop([0], axis=1)
```

```python
enc_df1.shape
```

```python
enc_df2
```

```python
enc_df2=enc_df2.drop([0], axis=1)
```

```python
enc_df2.shape
```

```python
df.shape
```

```python
df=pd.concat([df,enc_df1,enc_df2], axis=1, join='outer')
```

```python
df.shape
```

```python
df.head()
```

```python
df.shape
```

```python
X=df.iloc[:, 0:]
y=df.iloc[:,10]
X.drop(['Emp ID', 'Work_accident', 'salary', 'dept', 'status'], axis=1, inplace=True)
```

```python
X.head()
```

```python
X.shape
```

## 4. Estimating the accuracy

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```python
X_train.shape , X_test.shape , y_train.shape, y_test.shape
```

```python
clf_tree = DecisionTreeClassifier(criterion='entropy', max_depth=3, 
                                  random_state=17)
```

```python
clf_tree.fit(X_train, y_train)
```

```python
y_pred=clf_tree.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred)
```

### Thus, the proposed model "DECISION TREE ALGORITHM" can predict the employee churn with an accuracy of 94.37%
